**Contexto:** desenvolver o sistema GymTrack — um app para academias acompanharem treinos de alunos.

## Parte 1 — Classificação (RFs e RNFs)

**Requisitos Funcionais:**

* RF01 — O sistema deve validar o nome do exercício

* RF02 — O peso deve estar entre 1 e 300 kg

* RF03 — As repetições devem estar entre 1 e 50

---



**Requisitos Não Funcionais:**
* RNF01 — O registro deve ocorrer em menos de 200ms
* RNF02 — O sistema deve impedir o salvamento de dados se houver erros de validação.
* RNF03 — O sistema deve detectar o sistema operacional para garantir compatibilidade.


---



## Parte 2 — Código Python

```python
import time
import platform

print("🏋️ GymTrack — Validador de Requisitos")
print("=" * 50)

# --- DADOS DO TREINO (Mude aqui para testar as validações!) ---
exercicio = "Supino Reto"
peso_kg   = 250
repeticoes = 10

# Variável de controle para o RNF02 (Confiabilidade)
dados_validos = True

# -------------------------------------------------------
# PARTE 1: REQUISITOS FUNCIONAIS (RFs)
# -------------------------------------------------------

# RF01 — Validar nome do exercício
if exercicio != "":
    print("✅ [RF01] Exercício preenchido.")
else:
    print("❌ [RF01] ERRO: Nome do exercício não pode ser vazio!")
    dados_validos = False

# RF02 — Validar intervalo de peso
if 1 <= peso_kg <= 300:
    print("✅ [RF02] Peso dentro do limite permitido (1-300kg).")
else:
    print("❌ [RF02] ERRO: Peso fora do intervalo!")
    dados_validos = False

# RF03 — Validar quantidade de repetições
if 1 <= repeticoes <= 50:
    print("✅ [RF03] Repetições dentro do padrão (1-50).")
else:
    print("❌ [RF03] ERRO: Repetições inválidas!")
    dados_validos = False

print("-" * 50)

# -------------------------------------------------------
# PARTE 2: REQUISITOS NÃO FUNCIONAIS (RNFs)
# -------------------------------------------------------

# RNF01 — DESEMPENHO (Tempo de Resposta)
inicio = time.time()

# Simulação da ação de registro
if dados_validos:
    time.sleep(0.25) # Simula o delay de gravação
    print(f"🚀 REGISTRO REALIZADO: {exercicio} | {peso_kg}kg")
else:
    print("⚠️ REGISTRO NEGADO: O sistema bloqueou a gravação por erros nos dados.")

fim = time.time()
tempo_ms = (fim - inicio) * 1000

if tempo_ms < 200:
    print(f"✅ [RNF01] Desempenho: {tempo_ms:.1f}ms (Abaixo de 200ms)")
else:
    print(f"❌ [RNF01] Desempenho: {tempo_ms:.1f}ms (Lento demais!)")


# RNF02 — CONFIABILIDADE (Integridade)
# O sistema só deve processar se 'dados_validos' for True
if (not dados_validos and "NEGADO" in locals().get('_', '')) or dados_validos:
    print("✅ [RNF02] Confiabilidade: O sistema impediu falhas de integridade.")
else:
    print("❌ [RNF02] Confiabilidade: O sistema permitiu dados inválidos!")


# RNF03 — PORTABILIDADE (Compatibilidade de Ambiente)
os_name = platform.system()
if os_name in ["Linux", "Windows", "Darwin"]:
    print(f"✅ [RNF03] Portabilidade: Rodando em '{os_name}' (Ambiente homologado).")
else:
    print(f"❌ [RNF03] Portabilidade: Sistema operacional desconhecido.")

print("=" * 50)
print("Validação Concluída!")
```
