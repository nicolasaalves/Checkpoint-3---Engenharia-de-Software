## Parte 2 — Python no Colab/VS Code

Implementando o fluxo do diagrama em Python:

```python
def cadastro_usuario(email: str, senha: str, email_ja_existe: bool, confirmou_email: bool) -> str:
    """
    Implementação baseada no diagrama do Miro.

    Fluxo:
    1. Validar dados
    2. Verificar se e-mail é válido
    3. Se inválido -> exibir erro
    4. Se válido -> verificar duplicidade
    5. Se já cadastrado -> informar usuário
    6. Se não -> enviar e-mail de confirmação
    7. Aguardar confirmação do usuário
    8. Se confirmou -> finalizar cadastro
    """

    # 1) Validar dados básicos
    if email == "" or senha == "":
        return "Erro: e-mail e senha são obrigatórios."

    # 2) Validar formato do e-mail
    if "@" not in email or "." not in email:
        return "Erro: e-mail inválido."

    # 3) Verificar duplicidade
    if email_ja_existe:
        return "Usuário já cadastrado."

    # 4) Criar conta (simulação)
    conta_criada = True

    # 5) Enviar e-mail de confirmação (simulação)
    if conta_criada:
        print("E-mail de confirmação enviado para:", email)

    # 6) Aguardar confirmação
    if confirmou_email:
        return "Cadastro finalizado com sucesso! Acesso liberado."
    else:
        return "Cadastro pendente: confirme seu e-mail para liberar o acesso."


# Testes (não apague!)
print(cadastro_usuario("joao@email.com", "senha123", False, True))
print(cadastro_usuario("email-invalido", "senha123", False, True))
print(cadastro_usuario("joao@email.com", "senha123", True, True))
```
