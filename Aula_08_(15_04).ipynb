### **Justificativa dos Relacionamentos**

1️⃣ **Plataforma ◆── Catalogo
Tipo: Composição (losango preto)**

O catálogo pertence totalmente à plataforma.
Se a plataforma encerrar, o catálogo some junto.
Multiplicidade:

1 Plataforma -> 1 Catalogo

Ou seja:
Cada plataforma possui um catálogo.



---



2️⃣ **Catalogo ◇── Filme
Tipo: Agregação (losango branco)**

O filme pode existir sem aquele catálogo.
Pode sair da Netflix e ir para Prime.
Multiplicidade:

1 Catalogo -> * Filmes

Ou seja:
Um catálogo tem vários filmes.


---


3️⃣ **Usuario ◆── Avaliacao
Tipo: Composição**

A avaliação depende do usuário.
Se apagar usuário, avaliações somem.
Multiplicidade:

1 Usuario -> * Avaliações

Ou seja:
Um usuário pode fazer várias avaliações.


---


4️⃣ **Avaliacao ── Filme
Associação Simples**

Cada avaliação é feita para um filme.
Multiplicidade:

(*) Avaliações -> 1 Filme

Ou seja:
Um filme pode receber várias avaliações.

## 💻 Etapa 2 — Implementação das classes em Python seguindo o diagrama criado no Miro.

```python
# ===============================
# Classe Plataforma
# ===============================
class Plataforma:
    def __init__(self, nome, pais):
        self.nome = nome
        self.pais = pais
        self.catalogo = None

    def adicionar_catalogo(self, catalogo):
        self.catalogo = catalogo


# ===============================
# Classe Catalogo
# ===============================
class Catalogo:
    def __init__(self, titulo, qtdFilmes):
        self.titulo = titulo
        self.qtdFilmes = qtdFilmes
        self.filmes = []

    def add_filme(self, filme):
        self.filmes.append(filme)
        self.qtdFilmes += 1

    def listar_filmes(self):
        print(f"\nCatálogo: {self.titulo}")
        print("Filmes disponíveis:")
        for filme in self.filmes:
            print(f"- {filme.titulo} | {filme.genero} | {filme.duracao} min")


# ===============================
# Classe Filme
# ===============================
class Filme:
    def __init__(self, titulo, duracao, genero):
        self.titulo = titulo
        self.duracao = duracao
        self.genero = genero


# ===============================
# Classe Avaliacao
# ===============================
class Avaliacao:
    def __init__(self, nota, comentario):
        self.nota = nota
        self.comentario = comentario


# ===============================
# Classe Usuario
# ===============================
class Usuario:
    def __init__(self, nome, email, plano):
        self.nome = nome
        self.email = email
        self.plano = plano
        self.avaliacoes = []

    def avaliar(self, filme, avaliacao):
        self.avaliacoes.append((filme, avaliacao))

    def ver_avaliacoes(self):
        print(f"\nAvaliações de {self.nome}:")
        for filme, avaliacao in self.avaliacoes:
            print(f"Filme: {filme.titulo}")
            print(f"Nota: {avaliacao.nota}")
            print(f"Comentário: {avaliacao.comentario}")
            print("-" * 30)


# ===============================
# TESTE
# ===============================
netflix = Plataforma("Netflix", "EUA")

catalogo = Catalogo("Filmes em Destaque", 0)

filme1 = Filme("Oppenheimer", 180, "Drama")
filme2 = Filme("Barbie", 114, "Comédia")

catalogo.add_filme(filme1)
catalogo.add_filme(filme2)

netflix.adicionar_catalogo(catalogo)

usuario = Usuario("Ana", "ana@email.com", "Premium")

avaliacao = Avaliacao(9.5, "Incrível! Assisti duas vezes")

usuario.avaliar(filme1, avaliacao)

catalogo.listar_filmes()
usuario.ver_avaliacoes()
```
